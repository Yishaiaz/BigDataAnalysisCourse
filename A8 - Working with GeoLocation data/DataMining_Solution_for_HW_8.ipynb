**Question 1:** Select a book or a movie's subtitle, and find the locations that appear in the book/movie. Then, draw these locations on a map (at least 10 locations) (30pt). Connect the discovered locations with lines according to their order of appearance
in the book. For example, if location B appeared immediately after location A, then draw a line connecting locations A and B (20pt).

```python
def extract_locations_by_paragraph(book_txt: str) -> List:
  # extract all locations in book, locations are retrieved as an ordered list by order of appearance

  doc = nlp(book_txt)
  entities = doc.ents 
  all_locations_found = []
  for entity in entities:
    if entity.label_ == "GPE" or entity.label_ == 'LOC':
      stripped_entity = entity.text.lower().strip("_ \n\t.")
      # to remove any "locations" that are too short to probably be real locations (or informative)
      if len(stripped_entity) > 2:
        all_locations_found.append(stripped_entity)
  
  return all_locations_found

book_locations = None

with open('/content/datasets/books/The_Freaks_of_Mayfair.txt','r') as book_f:
  book_t = book_f.read()
  book_locations = extract_locations_by_paragraph(book_t)

# limit to 100 locations, remaining 46 unique locations
book_locations = book_locations[:100]
```

```python
# Get coordinates for each location, if no coordinates are found, replace the location value in list with None (to retain order of appearance proximity in the original text)
import time 
geolocator = Nominatim(user_agent="DataMiningHW8_BGU_Educational")

locations_coordinates = {}
for loc_idx, loc in tqdm(enumerate(book_locations)):
  # due to API constraints, we must wait at least a second between requests
  time.sleep(1)
  location = geolocator.geocode(loc)
  if location is None:
    book_locations[loc_idx] = None
  elif loc not in locations_coordinates:
    locations_coordinates[loc] = (location.latitude, location.longitude)
```

```python
import folium

# using folium interactive maps
folium_map = folium.Map()


# position all locations
for loc in tqdm(book_locations):
  if loc in locations_coordinates:
    loc_lat, loc_long = locations_coordinates[loc] 
    folium.Marker(
    location=[loc_lat, loc_long], # coordinates for the marker (Earth Lab at CU Boulder)
    popup=f'{loc.capitalize()}', # pop-up label for the marker
    icon=folium.Icon()).add_to(folium_map)

# plot connective lines between close locations in text (by appearance)
for loc_idx, loc in tqdm(enumerate(book_locations[:-1])):
  if book_locations[loc_idx] is not None and book_locations[loc_idx+1] is not None:
    points = [(locations_coordinates[book_locations[loc_idx]]), 
              (locations_coordinates[book_locations[loc_idx+1]])]
    folium.PolyLine(points).add_to(folium_map)

folium_map
```

**Question 3:** Select a request type from [San Francisco 311 dataset](https://www.kaggle.com/datasf/case-data-from-san-francisco-311).
Create an interactive map that somehow presents in which areas this request occurs the most (35pt).

```python
case_data_sframe = turi.SFrame.read_csv('/content/datasets/sf_311/Case_Data_from_San_Francisco_311__SF311.csv')
```

```python
# find how much from each request type we have:
all_requests_types_and_counts = case_data_sframe.groupby(key_column_names='Request Type', operations={'count': agg.COUNT()})
all_requests_types_and_counts.sort('count', ascending=False)
```

```python
chosen_request_type = 'Damaged Parking_Meter'
chosen_request_data_sframe = case_data_sframe[case_data_sframe['Request Type']==chosen_request_type]

def convert_point(x):
  try:
    return (float(x.strip('()').split(',')[0]), float(x.strip('()').split(',')[1]))
  except:
      print(x)
      
chosen_request_data_sframe['Point'] = chosen_request_data_sframe['Point'].apply(lambda x: convert_point(x))
```

```python
chosen_request_data_sframe['Point'][-2]
```

```python
def find_max_min(numbers, find_max, find_lt):
  if find_lt:
    to_find = np.array([x[0] for x in numbers])
  else:
    to_find = np.array([x[1] for x in numbers])
  if find_max:
    return to_find.max()
  return to_find.min()

chosen_request_type_data_by_neighboorhood = chosen_request_data_sframe.groupby(key_column_names='Neighborhood', operations={'count': agg.COUNT(), 'all_points': agg.CONCAT('Point')})
chosen_request_type_data_by_neighboorhood['neighborhood_address'] = chosen_request_type_data_by_neighboorhood['Neighborhood'].apply(lambda x: f'{x}, San Fransisco, United States of America')

# normalize count
chosen_request_type_data_by_neighboorhood['normalized_count'] = chosen_request_type_data_by_neighboorhood['count'].apply(lambda x: x/chosen_request_type_data_by_neighboorhood['count'].max())
# finding the center for each neighborhood
chosen_request_type_data_by_neighboorhood['max_lng'] = chosen_request_type_data_by_neighboorhood['all_points'].apply(lambda x: find_max_min(x, True, True))
chosen_request_type_data_by_neighboorhood['max_lt'] = chosen_request_type_data_by_neighboorhood['all_points'].apply(lambda x: find_max_min(x, True, False))
chosen_request_type_data_by_neighboorhood['min_lng'] = chosen_request_type_data_by_neighboorhood['all_points'].apply(lambda x: find_max_min(x, False, True))
chosen_request_type_data_by_neighboorhood['min_lt'] = chosen_request_type_data_by_neighboorhood['all_points'].apply(lambda x: find_max_min(x, False, False))
chosen_request_type_data_by_neighboorhood['neighborhood_center_lt'] = chosen_request_type_data_by_neighboorhood[['max_lt', 'min_lt']].apply(lambda x: np.array(list(x.values())).mean())
chosen_request_type_data_by_neighboorhood['neighborhood_center_lng'] = chosen_request_type_data_by_neighboorhood[['max_lng', 'min_lng']].apply(lambda x: np.array(list(x.values())).mean())
chosen_request_type_data_by_neighboorhood
```

```python
import folium

# using folium interactive maps
center_of_all_neighborhoods = [chosen_request_type_data_by_neighboorhood['neighborhood_center_lng'].to_numpy().mean(), chosen_request_type_data_by_neighboorhood['neighborhood_center_lt'].to_numpy().mean()]
folium_map = folium.Map(location=center_of_all_neighborhoods ,zoom_start=12)

scale = 50
# position all locations
for row in tqdm(chosen_request_type_data_by_neighboorhood):
  normalized_neighborhood_count = row['normalized_count']
  # color = (255 * normalized_neighborhood_count, 255 * normalized_neighborhood_count, 255 * normalized_neighborhood_count)
  color = "red"
  # if loc in locations_coordinates:
  loc = f'{row["Neighborhood"].capitalize()}, #Requests={row["count"]}'
  loc_long, loc_lat = row['neighborhood_center_lng'], row['neighborhood_center_lt']
  folium.CircleMarker(
  location=[loc_long, loc_lat],
    radius=normalized_neighborhood_count*scale,
    popup=loc,
    color=color,
    fill=True,
    fill_color=color,
).add_to(folium_map)


folium_map
```

Circles radii are relative to the amount of requests which occured in that area
